```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris = pd.read_excel("D:\ICTAK\dataset\iris.xls")
iris.head()
```

```python
iris.info()
```

```python
iris.mean()
```

```python
iris.isna().any(axis=0)
```

```python
plt.style.use('ggplot')
plt.figure(figsize=(8,8))
sns.scatterplot(data=iris,x ='SL', y="SW", hue='Classification')
plt.xlabel('sepal Length')
plt.ylabel('sepal Width')
plt.title('Sepal Length and Width of Iris species')
```

```python
plt.style.use('ggplot')
plt.figure(figsize=(8,8))
sns.scatterplot(data=iris,x ='PL', y="PW", hue='Classification')
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.title('Petal Length and Width of Iris species')
```

```python
plt.style.use('ggplot')
sns.displot(data=iris,x ='PW',hue='Classification',height=6.5,multiple='stack')
plt.title('Petal Width count of Iris species')
```
